- Find total immigrants from each category by area name and plot barchart for each area
- Plot total immigration trend(immigration of world over the years i.e. year vs Total count for that year) using line plot.
- Compare Immigration Trend for Asia, Africa and Europe AreaName (Use Line plot).
- Plot Total Immigrants based on DevelopmenRegion 

  
- Plot immigration trend for India using line plot
- Plot immigration Trend(year vs count) for the coutries with top 5 total immigrants. Use Line plot
- For the above use area plot with alpha = 0.5


- Compare Immigration Trend for top-5 countires using stacked area plot

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# - Plot bar chart of total no of immigrants for countires top 5 countires having max no of total immigrants

aus = pd.read_excel('AustraliaFiltered_Manas.xlsx')
max_imm = (aus.sort_values(by=['Total'], ascending=False)['Total']).head()
max_imm_ctry = (aus.sort_values(by=['Total'], ascending=False)['Country']).head()

plt.figure(figsize=(15,3))
plt.bar(max_imm_ctry,max_imm)
```

```python
#- Find total immigrants from each category by area name and plot barchart for each area
Area_Cat_Data = aus.groupby(by = ["AreaName"])
df = Area_Cat_Data.AreaName.unique() 
areaName = []
areaImm = []
for i in df:
    areaName.append(i[0])
    total = aus.loc[aus['AreaName'] == i[0],'Total'].sum()
    areaImm.append(total)

areaName = pd.Series(areaName)
areaImm = pd.Series(areaImm)

plt.figure(figsize=(14,3))
plt.bar(areaName, areaImm)
```

```python
#Plot total immigration trend(immigration of world over the years i.e. year vs Total count for that year) using line plot.
yr_world_data = aus.drop(['Unnamed: 0','Country', 'AreaName', 'Region', 'DevName', 'Total'], axis=1)
yr_world_data = yr_world_data.replace(to_replace='..', value=0)
yr_world_data = yr_world_data.sum()
yr_world_data_fin = []
for i in yr_world_data:
    yr_world_data_fin.append(i)

plt.figure(figsize=(14,5))
plt.plot(yr_world_data_fin, label = 'YOY Total')
plt.legend()
plt.show()
```

```python
#Compare Immigration Trend for Asia, Africa and Europe AreaName (Use Line plot).

Asia_Data = aus.loc[aus['AreaName'] == 'Asia']
Asia_Data = Asia_Data.replace(to_replace='..', value=0)
years = (Asia_Data.drop(['Unnamed: 0','Country', 'AreaName', 'Region', 'DevName', 'Total'], axis=1)).sum()
Asia_Data_fin = []
for i in years.index:
    Asia_Data_fin.append(years[i])
    
africa_data = aus.loc[aus['AreaName'] == 'Africa']
africa_data = africa_data.replace(to_replace='..', value=0)
years = (africa_data.drop(['Unnamed: 0','Country', 'AreaName', 'Region', 'DevName', 'Total'], axis=1)).sum()
africa_data_fin = []
for i in years.index:
    africa_data_fin.append(years[i])
    
eur_data = aus.loc[aus['AreaName'] == 'Europe']
eur_data = eur_data.replace(to_replace='..', value=0)
years = (eur_data.drop(['Unnamed: 0','Country', 'AreaName', 'Region', 'DevName', 'Total'], axis=1)).sum()
eur_data_fin = []
for i in years.index:
    eur_data_fin.append(years[i])


plt.figure(figsize=(14,5))
plt.plot(Asia_Data_fin, label = 'Asia')
plt.plot(africa_data_fin, label = 'Africa')
plt.plot(eur_data_fin, label = 'Europe')
plt.legend()
plt.show()
```

```python
#Plot Total Immigrants based on Development Region

developed_reg = aus[aus['DevName'] == "Developed regions"]['Total']
developing_reg = aus[aus['DevName'] == "Developing regions"]['Total'] 


plt.figure(figsize=(14,5))
plt.plot(developed_reg, label = 'Developed')
plt.plot(developing_reg, label = 'Developing')
plt.legend()
plt.show()
```

```python
# Plot immigration trend for India using line plot
ind_data = aus.loc[aus['Country'] == 'India']
ind_data = ind_data.replace(to_replace='..', value=0)
years = (ind_data.drop(['Unnamed: 0','Country', 'AreaName', 'Region', 'DevName', 'Total'], axis=1))
ind_data_fin = []
for i in years:
    ind_data_fin.append(years[i].values)

plt.plot(ind_data_fin)
```

```python
#Plot immigration Trend(year vs count) for the coutries with top 5 total immigrants. Use Line plot

max_imm_ctry = (aus.sort_values(by=['Total'], ascending=False)['Country']).head()
max_imm = (aus.sort_values(by=['Total'], ascending=False)).head()
max_imm = max_imm.replace(to_replace='..', value=0)
max_imm = max_imm.drop(['Unnamed: 0','Country', 'AreaName', 'Region', 'DevName', 'Total'], axis=1)
UK = (max_imm.values.tolist())[0]
NZ = (max_imm.values.tolist())[1]
US = (max_imm.values.tolist())[2]
CHHK = (max_imm.values.tolist())[3]
CH = (max_imm.values.tolist())[4]

plt.figure(figsize=(14,5))
plt.plot(UK, label = 'United Kingdom')
plt.plot(NZ, label = 'New Zealand')
plt.plot(US, label = 'United States')
plt.plot(CHHK, label = 'CH Hong Kong')
plt.plot(CH, label = 'China')
plt.legend()
plt.show()
```

```python
#For the above use area plot with alpha = 0.5

max_imm_ctry = (aus.sort_values(by=['Total'], ascending=False)['Country']).head()
max_imm = (aus.sort_values(by=['Total'], ascending=False)).head()
max_imm = max_imm.replace(to_replace='..', value=0)
max_imm = max_imm.drop(['Unnamed: 0','Country', 'AreaName', 'Region', 'DevName', 'Total'], axis=1)
UK = (max_imm.values.tolist())[0]
NZ = (max_imm.values.tolist())[1]
US = (max_imm.values.tolist())[2]
CHHK = (max_imm.values.tolist())[3]
CH = (max_imm.values.tolist())[4]

plt.figure(figsize=(14,5))
plt.plot(UK, label = 'United Kingdom', alpha=0.5)
plt.plot(NZ, label = 'New Zealand', alpha=0.5)
plt.plot(US, label = 'United States', alpha=0.5)
plt.plot(CHHK, label = 'CH Hong Kong', alpha=0.5)
plt.plot(CH, label = 'China', alpha=0.5)
plt.legend()
plt.show()
```

```python
#- Compare Immigration Trend for top-5 countires using stacked area plot

max_imm_ctry = (aus.sort_values(by=['Total'], ascending=False)['Country']).head()
max_imm = (aus.sort_values(by=['Total'], ascending=False)).head()
max_imm = max_imm.replace(to_replace='..', value=0)
max_imm = max_imm.drop(['Unnamed: 0','Country', 'AreaName', 'Region', 'DevName', 'Total'], axis=1)
UK = (max_imm.values.tolist())[0]
NZ = (max_imm.values.tolist())[1]
US = (max_imm.values.tolist())[2]
CHHK = (max_imm.values.tolist())[3]
CH = (max_imm.values.tolist())[4]

plt.figure(figsize=(14,5))
plt.fill_between(np.arange(34),UK, alpha=0.5, label = 'United Kingdom')
plt.fill_between(np.arange(34),NZ, alpha=0.5, label = 'New Zealand')
plt.fill_between(np.arange(34),US, alpha=0.5, label = 'United States')
plt.fill_between(np.arange(34),CHHK, alpha=0.5, label = 'CH Hong Kong')
plt.fill_between(np.arange(34),CH, alpha=0.5, label = 'China')
plt.legend()
plt.show()
```
