```python
import numpy as np
import pandas as pd
```

```python
# Read sheet "Australia by Residence" from Australia.xlsx
# Ignore the first 20 rows.
# List all Column and Row indexes(Print their names)

aus_data = pd.read_excel('Australia.xlsx',sheet_name ='Australia by Residence',skiprows = 20)
print(aus_data)

print()
print(list(aus_data.columns))

print()
print(aus_data.index)
```

```python
# Describe the dataset

aus_data.describe()
```

```python
# Use values and index attributes on both Row and Column Indexes
print(aus_data.columns.values)
print()
print(aus_data.index.values)
```

```python
# Check Shape, Datatypes and other attributes and observe the datatype of each columns' data

print(aus_data.shape)
print(aus_data.dtypes)
```

```python
# Delete all Data where the Type = "Emigrants" or Type is not "Immigrants" (Row Deletion)

drop_rows = aus_data[aus_data['Type'] == 'Emigrants'].index
drop1 = aus_data.drop(drop_rows, inplace = True)
print(drop1)
print(aus_data)
```

```python
# Find columns for which more than 50% data is missing and drop them (Column Deletion)

drop_columns = aus_data.columns[aus_data.isnull().mean() > 0.5]
drop = aus_data.drop(drop_columns, inplace = True)
print(drop)

print()
print(aus_data)
```

```python
# Drop Unnessary Columns: Type, Coverage, Area, Reg, Dev (Column deletion) Your columns should now look like "OdName,AreaName,RegName,DevName,1980 ... 2008"

drop_unwanted_columns = aus_data.drop(['Type', 'Coverage', 'AREA', 'REG', 'DEV'], axis = 1, inplace = True)
print(drop_unwanted_columns)
print(aus_data)
```

```python
# Rename columns :OdName -> Country; RegName -> Region

renaming = aus_data.rename(columns = {'OdName' : 'Country', 'RegName' : 'Region'}, inplace = True)
print(renaming)

print()
aus_data
```

```python
# Convert Column Names with Int type to Strings to avoid index errors with string indexing "1980, 1981 etc." => " '1980', '1981' ... " (Use rename)

aus_data.rename(columns = str, inplace =True)
print(list(aus_data))
```

```python
# Find Total number of immigrants for all years, and create a new column Total in data Frame

aus_data['Total'] = aus_data.iloc[:,4:39].sum(axis=1)

```

```python
# Sort Data by the new column Total in descending order and analyze the fist few rows. (get first few rows using head)

print(aus_data.sort_values(by='Total', ascending = False).head())
```

```python
# find something weird in the dataset ? Delete that row 'world'

drop_world = aus_data[aus_data['AreaName'] == 'World'].index
drop_world1 = aus_data.drop(drop_world, axis=0, inplace = True)
print(drop_world1)

print()
print(aus_data)
```

```python
# Analyze last few rows also in the sorted result (Don't sort inplace)

print(aus_data.sort_values(by='Total', ascending = False).tail())
```

```python
# Check if these have all unique values or not:

# countries
# continents (AreaName)
# country continent pairs

print(aus_data['Country'].is_unique)

print()
print(aus_data['AreaName'].is_unique)

print()
print(aus_data['Country'].index.is_unique)
```

```python
# Find countries for which no of immigrants for all years is 0

print(aus_data[aus_data['Total'] == 0].Country)
```

```python
# Find count of above countries

a = aus_data[aus_data['Total'] == 0].Country
print(a.count())
```

```python
# Find countries for which no of immigrants is 0 for any year

print(aus_data.Country[aus_data[aus_data.columns[4:]].apply(lambda x:x==0, axis =0).any(axis=1)])
```

```python
# Drop the row where AreaName is 'World'

drop_world1 = aus_data[aus_data['AreaName'] == 'World'].index
drop_world2 = aus_data.drop(drop_world1, axis=0, inplace = True)
print(drop_world2)

print()
print(aus_data)
```

```python
# Save the dataset to a new file AustraliaFiltered.xlsx

aus_data.to_excel('Karan_AustraliaFiltered.xlsx')
```
